```python
#Script for bid1
import os
import pdfplumber
from haystack.document_stores import FAISSDocumentStore
from haystack.nodes import EmbeddingRetriever, FARMReader
from haystack.pipelines import ExtractiveQAPipeline
from haystack.utils import print_answers
import json
from bs4 import BeautifulSoup

# Initialize the FAISS Document Store
document_store = FAISSDocumentStore(faiss_index_factory_str="Flat", sql_url="sqlite:///faiss_doc_store.db")

# Function to extract text from PDFs using pdfplumber
def extract_pdf_text(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Function to extract text from HTML
def extract_html_text(html_path):
    with open(html_path, 'r') as file:
        soup = BeautifulSoup(file, 'html.parser')
    return soup.get_text()

# Paths to your PDFs and HTML files
pdf_paths = ['/content/Addendum1.pdf', '/content/Addendum2.pdf', '/content/FinalDocument.pdf']  # Add your PDF paths here
html_file_path = '/content/Details.html'  # Add your HTML file path here

# Load and extract text from PDF documents
documents = []
for pdf_path in pdf_paths:
    pdf_text = extract_pdf_text(pdf_path)
    documents.append({"content": pdf_text, "meta": {"source": pdf_path}})

# Load and extract text from HTML document
html_text = extract_html_text(html_file_path)
documents.append({"content": html_text, "meta": {"source": html_file_path}})

# Write documents to the document store
document_store.write_documents(documents)

# Initialize the EmbeddingRetriever with a pre-trained model
retriever = EmbeddingRetriever(document_store=document_store, embedding_model="deepset/sentence_bert")

# Update FAISS embeddings
document_store.update_embeddings(retriever)

# Initialize FARMReader with a model for question-answering
reader = FARMReader(model_name_or_path="deepset/roberta-base-squad2")

# Define the pipeline combining the retriever and reader
pipeline = ExtractiveQAPipeline(reader, retriever)

# List of fields to extract
fields = [
    "Bid Number",
    "Title",
    "Due Date",
    "Bid Submission Type",
    "Term of Bid",
    "Pre Bid Meeting",
    "Installation",
    "Bid Bond Requirement",
    "Delivery Date",
    "Payment Terms",
    "Any Additional Documentation Required",
    "MFG for Registration",
    "Contract or Cooperative to use",
    "Model_no",
    "Part_no",
    "Product",
    "contact_info",
    "company_name",
    "Bid Summary",
    "Product Specification",
    "Value"
]

# Dictionary to store extracted data
extracted_data = {}

# Loop through each field and ask the model to extract information
for field in fields:
    question = f"What is the {field} in this document?"
    result = pipeline.run(query=question, params={"Retriever": {"top_k": 1}, "Reader": {"top_k": 1}})

    # If the result contains answers, extract the answer, else mark as "Not found"
    if result['answers']:
        extracted_data[field] = result['answers'][0].answer
    else:
        extracted_data[field] = "Not found"

# Print the extracted data
print("Extracted Data:")
for field, answer in extracted_data.items():
    print(f"{field}: {answer}")

#save the results to a JSON file
output_path = 'extracted_data.json'  # Define the path to save the JSON file
with open(output_path, 'w', encoding="utf-8") as json_file:
    json.dump(extracted_data, json_file, indent=4)

# Print the path of the output JSON file
print(f"Extracted data saved to: {output_path}")
```

```python
#Script for bid 2
import os
import pdfplumber
from bs4 import BeautifulSoup
from haystack.document_stores import FAISSDocumentStore
from haystack.nodes import EmbeddingRetriever, FARMReader
from haystack.pipelines import ExtractiveQAPipeline
import json

# Initialize the FAISS Document Store
document_store = FAISSDocumentStore(faiss_index_factory_str="Flat", sql_url="sqlite:///faiss1_doc_store.db")

# Function to extract text from PDFs using pdfplumber
def extract_text_from_pdf(pdf_path):
    text = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                text += page.extract_text() or ""
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
    return text

# Function to extract text from HTML using BeautifulSoup
def extract_text_from_html(html_path):
    try:
        with open(html_path, "r", encoding="utf-8") as file:
            soup = BeautifulSoup(file, "html.parser")
        return soup.get_text()
    except Exception as e:
        print(f"Error processing {html_path}: {e}")
        return ""

# Paths to your documents
pdf_paths = [
    "/content/Contract_Affidavit.pdf",
    "/content/Dell_Laptop_Specs.pdf",
    "/content/Mercury_Affidavit.pdf",
    "/content/PORFP_-_Dell_Laptop_Final.pdf"
]
html_path = "/content/Dell Laptops w_Extended Warranty - Bid Information - {3} _ BidNet Direct.html"

# Extract text from all documents
documents = []
for pdf_path in pdf_paths:
    content = extract_text_from_pdf(pdf_path)
    documents.append({"content": content, "meta": {"source": pdf_path}})

html_content = extract_text_from_html(html_path)
documents.append({"content": html_content, "meta": {"source": html_path}})

# Write documents to the FAISS Document Store
document_store.write_documents(documents)

# Initialize the Embedding Retriever
retriever = EmbeddingRetriever(document_store=document_store, embedding_model="deepset/sentence_bert")

# Update embeddings in the FAISS Document Store
print("Updating embeddings...")
document_store.update_embeddings(retriever)

# Check synchronization
if document_store.get_document_count() != document_store.get_embedding_count():
    print("Mismatch detected! Reinitializing FAISS index...")
    document_store.faiss_index.reset()
    document_store.update_embeddings(retriever)

# Initialize FARMReader
reader = FARMReader(model_name_or_path="deepset/roberta-base-squad2")

# Define the Extractive QA pipeline
pipeline = ExtractiveQAPipeline(reader, retriever)

# Fields to extract
fields = [
    "Bid Number",
    "Title",
    "Due Date",
    "Bid Submission Type",
    "Term of Bid",
    "Pre Bid Meeting",
    "Installation",
    "Bid Bond Requirement",
    "Delivery Date",
    "Payment Terms",
    "Any Additional Documentation Required",
    "MFG for Registration",
    "Contract or Cooperative to use",
    "Model_no",
    "Part_no",
    "Product",
    "contact_info",
    "company_name",
    "Bid Summary",
    "Product Specification",
    "Value"
]

# Extract answers for each field
extracted_data = {}
for field in fields:
    query = f"What is the {field} in this document?"
    try:
        result = pipeline.run(query=query, params={"Retriever": {"top_k": 1}, "Reader": {"top_k": 1}})
        answer = result["answers"][0].answer if result["answers"] else "Not found"
        extracted_data[field] = answer
    except Exception as e:
        print(f"Error processing query for field '{field}': {e}")
        extracted_data[field] = "Error during extraction"

# Save extracted data to JSON
output_path = "aextracted_data.json"
with open(output_path, "w", encoding="utf-8") as json_file:
    json.dump(extracted_data, json_file, indent=4)

print(f"Extracted data saved to: {output_path}")
```
